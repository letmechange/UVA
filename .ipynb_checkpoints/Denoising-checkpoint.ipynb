```python
from voicefixer import VoiceFixer
import librosa
import soundfile as sf
import matplotlib.pyplot as plt

# Load audio
y, sr = librosa.load("voice file/Bertha1965_4_bear.wav", sr=44100)  # VoiceFixer requires 44100 Hz
sf.write("Output/denoised_Bertha1965_4_bear.wav", y, 44100)

# Plot pitch before
f0_before, _, _ = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))
plt.plot(f0_before, label="Pitch Before Denoising")
plt.legend()
plt.show()

# Run VoiceFixer (MAIN  FUNCTION)
model = VoiceFixer()
model.restore(input="voice file/Bertha1965_4_bear.wav", output="Output/denoised_Bertha1965_4_bear.wav")

# Load denoised audio
y_denoised, _ = librosa.load("Output/denoised_Bertha1965_4_bear.wav", sr=44100)

# Plot pitch after
f0_after, _, _ = librosa.pyin(y_denoised, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))
plt.plot(f0_after, label="Pitch After Denoising", color='orange')
plt.legend()
plt.show()
```
